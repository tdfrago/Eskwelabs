# Data Visualization using Seaborn

- Seaborn is a Python visualization library based on matplotlib. 
- It provides a high-level interface for drawing - attractive statistical graphics. Here is the [documentation](http://seaborn.pydata.org/index.html).
- Go ahead and import seaborn. By convention, we alias seaborn as sns


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("regional_stats.csv")
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.columns
```

### 1. Boxplot

Let's try to see if there's a difference among the scores of students in the National Achievement Test.

```python
nat = df[['NAT_MATH_PRIMARY', 'NAT_ENGLISH_PRIMARY', 'NAT_SCIENCE_PRIMARY',
          'NAT_MATH_SECONDARY', 'NAT_ENGLISH_SECONDARY', 'NAT_SCIENCE_SECONDARY']]
```

```python
plt.figure(figsize=(12,6))

sns.boxplot(x="value", y="variable", data=pd.melt(nat))
plt.title("Distribution of Scores in the National Achievement Test")

plt.show()
```

### 2. Correlation Plot

```python
df.corr()
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(df.corr(), center=0.00)

plt.show()
```

```python
plt.figure(figsize=(10,8))

sns.pairplot(df[['NET_ENROLMENT_ELEMENTARY', 'COHORT_SURVIVAL_ELEMENTARY',
                 'GROSS_ENROLLMENT_PRIMARY', 'TEACHER_PRIMARY', 'TEXTBOOKS_DISTRIBUTED']])

plt.show()
```
