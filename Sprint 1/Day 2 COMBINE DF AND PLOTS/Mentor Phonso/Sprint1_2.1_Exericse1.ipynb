12. Which school_id has the highest teacher-student ratio?

```python
schools
```

```python
new_df = schools.groupby("school_id")[['Total_Inst', 'enrollment']].agg({
    'Total_Inst':'first', 'enrollment':'sum'
})
```

```python
new_df['ratio'] = new_df['Total_Inst'] / new_df['enrollment']
```

```python
new_df.sort_values(by='ratio', ascending= False)
```

```python
df = schools.groupby('school_id')[['Total_Inst', 'enrollment']].agg({'Total_Inst':'first', 
                                                               'enrollment':'sum'})
df['ratio'] = df['Total_Inst'] / df['enrollment']
```

```python
df.sort_values(by='ratio', ascending=False)
```
